```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/kaggle/input/skripsi-data-ready-eda/data_ready_eda.csv')
data
```

```python
zero_value = data[data['standard_value'] == 0 ]
zero_value
```

```python
data_final = data[data['standard_value'] != 0]
data_final
```

```python
data_final.describe()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(data_final['standard_value'])

plt.ylabel(r"Nilai IC$_{50}$")
plt.title(r"Boxplot IC$_{50}$")

plt.show()
```

```python
# Hitung Q1, Q3, dan IQR
Q1 = data_final['standard_value'].quantile(0.25)
Q3 = data_final['standard_value'].quantile(0.75)
IQR = Q3 - Q1

# Tentukan batas bawah dan atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(lower_bound)
print(upper_bound)

# Cari outlier
outliers = data_final['standard_value'][(data_final['standard_value'] < lower_bound) | (data_final['standard_value'] > upper_bound)]
print("Outliers:", outliers)
```

```python
dataclean = data_final[(data_final['standard_value'] >= Q1) & (data_final['standard_value'] <=Q3) ]
dataclean
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(dataclean['standard_value'])

plt.ylabel(r"Nilai IC$_{50}$")
plt.title(r"Boxplot IC$_{50}$")

plt.show()
```

```python
dataclean['standard_value'].describe()
```

```python
print('Skewness: ',dataclean['standard_value'].skew())
print('Kurtosis: ',dataclean['standard_value'].kurtosis())
```

Skewness >1 maka akan ditransformasikan menggunakan log

```python
dataclean['standard_value']
```

```python
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.histplot(dataclean['standard_value'], bins=50, kde=True, color='skyblue')
plt.title(r'Distribusi Nilai IC$_{50}$')
plt.xlabel(r'IC$_{50}$')
plt.ylabel('Frekuensi')
plt.show()
```

```python
def calculate_pIC50(data):
    pIC50_values = []

    for value in data['standard_value']:
        molar_value = value * 1e-9  
        pIC50_values.append(-np.log10(molar_value))

    data['pIC50'] = pIC50_values
    data.drop(columns=['standard_value'], inplace=True)
    return data
```

1. Nilai diubah dari nM -> M (molar), makannya dikali 1e-9
2. pIC50=−log10(IC50 dalam molar) → Nilai lebih tinggi menunjukkan potensi yang lebih kuat (karena IC50 kecil = senyawa lebih aktif).

```python
data_pic50 = calculate_pIC50(dataclean)
data_pic50
```

```python
pd.set_option('display.max_colwidth', None)
check_data = data_pic50[round(data_pic50['pIC50'], 6)==6.996927]
check_data
```

```python
print('Skewness: ',data_pic50['pIC50'].skew())
print('Kurtosis: ',data_pic50['pIC50'].kurtosis())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data_pic50['pIC50'], bins=50, kde=True, color='skyblue')
plt.title(r'Distribusi Nilai pIC$_{50}$')
plt.xlabel(r'pIC$_{50}$')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(data_pic50['pIC50'])

plt.ylabel(r"pIC$_{50}$ Value")
plt.title(r"Boxplot pIC$_{50}$")

plt.show()
```
