# Search Target 

## Search Organism

Mencari organisme target dengan nama Plasmodium falciparum (parasit malaria)

```python
organism_target = new_client.target
organism_target_query = organism_target.search('Plasmodium falciparum')
list_targets = pd.DataFrame.from_dict(organism_target_query)
pd.set_option('display.max_rows', None)
list_targets
```

Memilih organisme target dengan id 0

```python
selected_target_falciparum = list_targets.target_chembl_id[0]
```

## Filter (IC50 and nM)

Memfilter aktivitas senyawa dengan tipe IC50 dan satuan nM

```python
activity = new_client.activity
res_falciparum = activity.filter(target_chembl_id=selected_target_falciparum)
```

```python
df_falciparum = pd.DataFrame.from_dict(res_falciparum)
```

```python
df_falciparum.to_csv('plasmodiumfalciparum_data.csv', index=False)
#df_vivax.to_csv('plasmodiumvivax_data.csv', index=False)
```
